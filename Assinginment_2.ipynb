```python
# 4. Write a Python program to solve quadratic equation?
import cmath
a = int(input())
b = int(input())
c = int(input())
# (-b ± (b ** 2 - 4 * a * c) ** 0.5) / (2 * a)
d = (b**2) - 4*a*c

ans1 = (-b-cmath.sqrt(d))/2*a
ans2 = (+b+cmath.sqrt(d))/2*a

print("solution are {},{}".format(ans1,ans2))
```
